```python
import ROOT
```

```python
from ROOT import RooNovosibirsk, RooRealVar, RooDataSet, RooPlot, RooArgSet
from ROOT.RooFit import Title
```

```python
x = RooRealVar('x', 'x', -10, 1)
mean = RooRealVar('mean','mean', .4, -10, 10)
sigma = RooRealVar('sigma','sigma', .6, 0, 1)
tail = RooRealVar('tail','tail', 1.1, 0, 3)
```

```python
novo = RooNovosibirsk('novo', 'novo', x, mean, sigma, tail)
```

```python
xframe = x.frame(Title("Novo PDF"))
```

```python
novo.plotOn(xframe)
```

```python
c1 = ROOT.TCanvas('c1','c1')
xframe.Draw()
c1.Draw()
```

```python
ds = novo.generate(RooArgSet(x), 1000)
```

```python
result = novo.fitTo(ds)
```

```python
xframe2 = x.frame(Title("Novo Fit PDF"))
ds.plotOn(xframe2)
novo.plotOn(xframe2)
c1 = ROOT.TCanvas('c1','c1')
xframe2.Draw()
c1.Draw()
```
